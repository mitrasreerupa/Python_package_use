```python
import urllib3
http = urllib3.PoolManager()
url = "https://zetcode.com/python/urllib3/"
try:
    resp = http.request('GET', url, timeout = 5.0, retries = 5, redirect = False)
except urllib3.exceptions.NewConnectionError:
    print('Connection failed')
print(resp.status)
print(resp.data.decode('utf-8'))
```
